### 예측 (2021년 기준, 가을은 2020년 기준)

PM은 대전(한국), JPM은 지난, TPM은 텐진, BPM은 베이징 미세먼지를 의미함.

1. TPM, BPM(초록색과 빨간색)은 오르고 내리는 패턴이 거의 같다.
2. 국내 미세먼지는 JPM(노란색)보다 TPM, BPM(초록색과 빨간색)과 더 상관관계가 높다.
3. 국내 미세먼지는 TPM과 BPM(초록색과 빨간색)과는 거의 변화 방향이 비슷하며, 하루 정도 뒤처지는 경향이 있다.(0~1일)
    - 여름은 0-1일, 겨울은 0-1일, 가을은 1-2일, 봄은 2일 정도 뒤처짐.
4. JPM(노란색)은 온도가 높을 때, TPM, BPM(초록색과 빨간색)보다 하루, 이틀 정도 앞서는 경향이 있다.
    -  JPM이 단독으로 패턴이 다르게 나타나는데, TPM과 BPM의 1~2일 전 패턴을 보면 비슷함. (여름)
    -  하지만 TPM과 BPM, 한국과 비교하였을 때는 다른 패턴이 종종 보여 그저 앞선다고 확신 불가능.
    -  가끔은 TPM과 BPM과 같은 날에 동일한 패턴을 보일 때도 있음. 
        - (많은 순서 : 겨울 > 가을 = 봄 > 여름) 
5. 온도가 높은 경우보다 낮은 경우에 중국 미세먼지에 비해 한국 미세먼지 변화량이 크지 않다. 
    - (변화량 : 가을 < 겨울 < 봄 < 여름)
    - 가을과 겨울의 경우, 봄과 여름에 비하여 중국에게 큰 영향을 받지 않음.

결론 : JPM, 즉, 지난이 가장 안쪽에 위치, 지난이 하루 뒤에 BPM, TPM인 베이징과 텐진에 영향을 주고,
베이징과 텐진이 하루 뒤에 한국의 대기질 PM에 영향을 줄 것으로 예상.
- 지도를 확인해본 결과, 지난이 가장 왼쪽에 위치하였고, 그 뒤로 텐진, 상하이, 대한민국으로 이어져 결론과 같은 결과가 나옴. 단, 텐진과 지난은 경도 기준으로는 차이가 거의 나지 않기에, 원인을 확정지을 수 없음. 
     - 지난과 텐진,베이징의 관계로 분석해보았을 때 거리 상으로는 대한민국이 더 나중에 효과가 나타나야하는데 1일 정도밖에 차이가 나지 않기 때문.
