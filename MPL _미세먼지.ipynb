## 데이터 합치기

```python
# 구분 쉽게 columns 이름 바꿔놓기

data_T.rename(columns = {'PM2.5':'TPM2.5'}, inplace = True)
data_B.rename(columns = {'PM2.5':'BPM2.5'}, inplace = True)
data_J.rename(columns = {'PM2.5':'JPM2.5'}, inplace = True)
data_T.rename(columns = {'PM10':'TPM10'}, inplace = True)
data_B.rename(columns = {'PM10':'BPM10'}, inplace = True)
data_J.rename(columns = {'PM10':'JPM10'}, inplace = True)
```

```python
data = data_T.merge(data_J)
data = data.merge(data_B)

data
```
