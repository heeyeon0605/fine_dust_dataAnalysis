```python
monthly = newdf.resample('M').sum()

monthly
```

```python
from statsmodels.tsa.stattools import adfuller

def df_test(time_series):
    
    print("Results of Dickey-Fuller Test:")
    dftest = adfuller(time_series, autolag='AIC')
    dfoutput = pd.Series(dftest[0:4], index=['Test Statistic', 'p-value',
                                            '#Lags Used', 
                                             'Number of Observations Used'])
    
    for key, value in dftest[4].items():
        dfoutput['Critical Value (%s)' %key] = value
        
    print(dfoutput)
```

```python
df_test(monthly.PM10)
```

```python
first_diff = monthly.PM10.diff()[1:]

print(list(first_diff))
```

```python
from datetime import timedelta

train_end = datetime(2020,12,31)
test_end = datetime(2021,6,30)

train_data = monthly[:train_end]
test_data = monthly[train_end + timedelta(days=1):test_end]
```

```python
test_data.tail(5)
```

```python
from pmdarima import auto_arima

# pmdarima.arima.ndiffs() # finding the best d and D
```

```python
from statsmodels.tsa.api import ExponentialSmoothing

air = ExponentialSmoothing(np.asarray(train_data['PM10']),
                                  seasonal_periods=12, trend='add',
                                  seasonal='add')

air_fit = air.fit()

air_forecast = air_fit.forecast(len(test_data))

plt.figure(figsize=(16,8))

plt.plot(train_data.index, train_data['PM10'], "b.-", label='Train Data')
plt.plot(test_data.index, test_data['PM10'], "ro-", label='Original Test Data')
plt.plot(test_data.index, air_forecast, "gx-", label='Air Forecast Data')
plt.ylabel('PM10', fontsize=16)
plt.xlabel('date', fontsize=16)
plt.legend(loc='best')
plt.title('Air Forecast', fontsize=10)
plt.show()
```

### Sarima

```python
Sarimax_model = auto_arima(train_data.PM10,
                          start_p=0,
                          start_q=0,
                          max_p=3,
                          max_q=3,
                          m=12,
                          test='adf',
                          seasonal=True,
                          d=1,
                          D=1,
                          trace=True,
                          error_action='ignore',
                          suppress_warnings=True,
                          stepwise=True)

Sarimax_model.summary()
```

```python
from statsmodels.tsa.statespace.sarimax import SARIMAX

my_order = (1,1,0)

my_seasonal_order = (2, 1, 1, 12)

# define model
model = SARIMAX(train_data.PM10, order=my_order, seasonal_order=
                my_seasonal_order)
```

```python
# fit the model
model_fit = model.fit()
```

```python
# get the predictions and residuals
predictions = model_fit.forecast(len(test_data.PM10))
predictions = pd.Series(predictions, index=test_data.index)
residuals = test_data.PM10 - predictions
```

```python
plt.figure(figsize=(10,4))
plt.plot(residuals)
plt.axhline(0, linestyle='--', color='k')
plt.title('Residuals from SARIMA Model', fontsize=20)
plt.ylabel('Error', fontsize=16)
```

```python
from sklearn.metrics import mean_squared_error, r2_score

# 평가 지표 계산
mse = mean_squared_error(test_data['PM10'], predictions)
rmse = np.sqrt(mse)

r2 = r2_score(test_data['PM10'], predictions)

print("MSE: {0:.3f}, RMSE: {1:.3f}".format(mse, rmse))
print("R2: {:3f}".format(r2))
```

```python
# 평가 지표 계산
mse2 = mean_squared_error(test_data['PM10'], air_forecast)
rmse2 = np.sqrt(mse2)

r22 = r2_score(test_data['PM10'], air_forecast)

print("MSE: {0:.3f}, RMSE: {1:.3f}".format(mse2, rmse2))
print("R2: {:3f}".format(r22))
```
