```python
period_idx = pd.period_range('2017-06-01', '2021-06-30', freq='D')
sum = 0

for day in period_idx : 
    sum = sum + 1
    
sum # 본 데이터의 row 수가 1490 인것을 보아 하루가 모자란 것을 알 수 있음
```
